```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('general_data.csv')
df
```

```python
inputs = df.drop('Over18',axis='columns')
```

```python
from sklearn.preprocessing import LabelEncoder
le_Attrition = LabelEncoder()
le_BusinessTravel = LabelEncoder()
le_Department = LabelEncoder()
le_EducationField = LabelEncoder()
le_Gender = LabelEncoder()
le_JobRole= LabelEncoder()
le_Age = LabelEncoder()
le_DistanceFromHome = LabelEncoder()
le_Education = LabelEncoder()
le_EmployeeCount = LabelEncoder()
le_MonthlyIncome = LabelEncoder()
le_NumCompaniesWorked= LabelEncoder()
le_PercentSalaryHike = LabelEncoder()
le_StandardHours = LabelEncoder()
le_TrainingTimesLastYear = LabelEncoder()
le_YearsAtCompany = LabelEncoder()
le_YearsSinceLastPromotion = LabelEncoder()
le_YearsWithCurrManager= LabelEncoder()
le_JobLevel=LabelEncoder()
```

```python
inputs['Attrition_n'] = le_Attrition.fit_transform(inputs['Attrition'])
inputs['BusinessTravel_n'] = le_BusinessTravel.fit_transform(inputs['BusinessTravel'])
inputs['Department_n'] = le_Department.fit_transform(inputs['Department'])
inputs['EducationField_n'] = le_EducationField.fit_transform(inputs['EducationField'])
inputs['Gender_n'] = le_Gender.fit_transform(inputs['Gender'])
inputs['JobRole_n'] = le_JobRole.fit_transform(inputs['JobRole'])
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['DistanceFromHome_n'] = le_DistanceFromHome.fit_transform(inputs['DistanceFromHome'])
inputs['Education_n'] = le_Education.fit_transform(inputs['Education'])
inputs['EmployeeCount_n'] = le_EmployeeCount.fit_transform(inputs['EmployeeCount'])
inputs['MonthlyIncome_n'] = le_MonthlyIncome.fit_transform(inputs['MonthlyIncome'])
inputs['NumCompaniesWorked_n'] = le_NumCompaniesWorked.fit_transform(inputs['NumCompaniesWorked'])
inputs['PercentSalaryHike_n'] = le_PercentSalaryHike.fit_transform(inputs['PercentSalaryHike'])
inputs['StandardHours_n'] = le_StandardHours.fit_transform(inputs['StandardHours'])
inputs['TrainingTimesLastYear_n'] = le_TrainingTimesLastYear.fit_transform(inputs['TrainingTimesLastYear'])
inputs['YearsAtCompany_n'] = le_YearsAtCompany.fit_transform(inputs['YearsAtCompany'])
inputs['YearsSinceLastPromotion_n'] = le_YearsSinceLastPromotion.fit_transform(inputs['YearsSinceLastPromotion'])
inputs['YearsWithCurrManager_n'] = le_YearsWithCurrManager.fit_transform(inputs['YearsWithCurrManager'])
inputs['JobLevel_n'] = le_JobLevel.fit_transform(inputs['JobLevel'])
```

```python
inputs
```

```python
inputs_n=inputs.drop(['Attrition','BusinessTravel','Department','EducationField','Gender','MaritalStatus','JobRole','Age','DistanceFromHome','Education','EmployeeCount','JobLevel','MonthlyIncome','NumCompaniesWorked','PercentSalaryHike','StandardHours','JobLevel','TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager',], axis='columns')
```

```python
inputs_n
```

```python
y=inputs_n.Attrition_n
```

```python
X=inputs_n[['Age_n','DistanceFromHome_n','Education_n','EmployeeCount_n','JobLevel_n','MonthlyIncome_n','NumCompaniesWorked_n','PercentSalaryHike_n','StandardHours_n','TrainingTimesLastYear_n','YearsAtCompany_n','YearsSinceLastPromotion_n','YearsWithCurrManager_n','BusinessTravel_n','Department_n','EducationField_n','Gender_n','JobRole_n']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test)
```
