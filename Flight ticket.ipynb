```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  
```

```python
df = pd.read_csv('Data_Train.csv')
df.head(50)
```

```python
df['Additional_Info'].value_counts()
```

```python
df['Duration'].value_counts()
```

```python
df.replace({'2 Long layover': 'Long layover'},inplace=True)
df.replace({'1 Short layover': 'Short layover'},inplace=True)
df.replace({'1 Long layover': 'Long layover_n'},inplace=True)
```

```python
N_df = df.drop(['Date_of_Journey'],axis='columns')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Airline =LabelEncoder()
le_Source = LabelEncoder()
le_Destination = LabelEncoder()
le_Total_Stops= LabelEncoder()
le_Additional_Info=LabelEncoder
```

```python
N_df['Airline_n'] = le_Airline.fit_transform(N_df['Airline'])
N_df['Source_n'] = le_Source.fit_transform(N_df['Source'])
N_df['Destination_n'] = le_Destination.fit_transform(N_df['Destination'])
N_df['Additional_Info_n'] = le_Airline.fit_transform(N_df['Additional_Info'])
N_df['Duration_n'] = le_Airline.fit_transform(N_df['Duration'])
```

```python
N_df.head()
```

```python
dum=pd.get_dummies(N_df.Total_Stops)
dum.head()
```

```python
F_df= pd.concat([N_df,dum],axis='columns')
F_df.head()
```

```python
FF_df=F_df.drop(['Airline','Source','Destination','Route','Dep_Time','Arrival_Time','Total_Stops'],axis='columns')
```

```python
X=FF_df[['Airline_n','Source_n','Destination_n','1 stop','2 stops','3 stops', '4 stops', 'non-stop','Additional_Info_n']]
```

```python
XX=FF_df[['Airline_n','Source_n','Destination_n','1 stop','2 stops','3 stops', '4 stops', 'non-stop','Price']]
X.head(50)
```
